# Get Connect stats

```python
from datetime import date, timedelta
import pandas as pd

# Define your target end date
end_date = date(2023, 10, 30)

# Calculate the start date as 2 weeks (14 days) before the end date
start_date = end_date - timedelta(days=13)

datelist = []

current_date = start_date
while current_date <= end_date:
    datelist.append(current_date)
    current_date += timedelta(days=1)

datelist_new = [x.isoformat() for x in datelist]

dataframes_list = []  # List to store each day's DataFrame

for day in datelist_new:
    current_entry = garmin.get_sleep_data(day)
    current_entry_df = pd.DataFrame([current_entry])  # Convert the dictionary to a DataFrame
    dataframes_list.append(current_entry_df)

# Concatenate all the DataFrames in the list
week_sleep_df = pd.concat(dataframes_list, ignore_index=True)

week_sleep_df
```

```python
qualifier_dict = {"nan": 0,
                 "FAIR": 1,
                 "EXCELLENT": 2}
def encoder(data, converter_dict):
    if data in converter_dict:
        return converter_dict[data]
    else:
        return None
```

```python
modified_list = ['sleepScores.totalDuration.qualifierKey', 'sleepScores.stress.qualifierKey',
          'sleepScores.awakeCount.qualifierKey', 'sleepScores.overall.qualifierKey', 'sleepScores.remPercentage.qualifierKey',
          'sleepScores.lightPercentage.qualifierKey', 'sleepScores.deepPercentage.qualifierKey', 'sleepScores.restlessness.qualifierKey']

for column in modified_list:
    expanded_df[column] = expanded_df[column].apply(encoder, args=(qualifier_dict,))
```

```python
import matplotlib.colors as mcolors
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

colors = ['darkblue', 'blue', 'skyblue', 'salmon', 'darkred']
cmap = mcolors.ListedColormap(colors)

to_drop = ['calendarDate','sleepWindowConfirmationType', "ageGroup", "sleepScoreFeedback", "sleepScoreInsight",
          'sleepScorePersonalizedInsight']
corr_matrix = expanded_df.drop(to_drop, axis='columns').corr()
to_drop = corr_matrix.columns[corr_matrix.isna().all()].tolist()
corr_matrix = corr_matrix.drop(to_drop, axis=0)
corr_matrix = corr_matrix.drop(to_drop, axis=1)

# Get correlations greater than 0.7
# The stack() function reshapes the filtered DataFrame by stacking the
# columns into a single column, resulting in a Series.
correlations = corr_matrix[corr_matrix > 0.8].stack().reset_index()
correlations.columns = ['Feature 1', 'Feature 2', 'Correlation']

# Filter out self-correlations
correlations = correlations[correlations['Feature 1'] != correlations['Feature 2']]

# handle duplicates like [Feature 1 Feature 2] and [Feature 2 Feature 1]
correlations[['Feature 1', 'Feature 2']] = np.sort(correlations[['Feature 1', 'Feature 2']], axis=1)
correlations = correlations.drop_duplicates(subset=['Feature 1', 'Feature 2'])

# Sort correlations in descending order
correlations = correlations.sort_values(by='Correlation', ascending=False)

fig, ax = plt.subplots(figsize=(10, 8))
sb.heatmap(corr_matrix, cmap=cmap)
plt.title(f'Correlation Matrix - expanded_df')

#print(f'Correlations > 0.7 in expanded_df:')
#print(correlations)

plt.show()
```
